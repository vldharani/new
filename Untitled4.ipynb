```python
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed (10)
data = pd.DataFrame({
'value': np.concatenate ([np.random.normal(0, 1, 100), np.random.normal(10, 1, 10)])
})
Q1 = data['value'].quantile (0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound= Q3 + 1.5 * IQR

outliers = data[(data['value'] < lower_bound) |(data['value'] > upper_bound)]
print(f"Outliers based on Box Plot criteria: \n{outliers}")
plt.figure(figsize=(12,6))
plt.subplot(1, 2, 1)
sns.boxplot(x=data['value'])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


np.random.seed(10)
data = pd.DataFrame({
    'value': np.concatenate([np.random.normal(0, 1, 100), np.random.normal(10, 1, 10)])
})

Q1 = data['value'].quantile(0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

data['outlier'] = (data['value'] < lower_bound) | (data['value'] > upper_bound)

plt.figure(figsize=(10, 6))
sns.boxplot(x=data['value'])
plt.title("Box Plot with Outliers")
plt.show()

print("Outliers:")
print(data[data['outlier']])
```

```python
def calculate_fare(base_price, rate_per_km, distance):
    return base_price + (rate_per_km * distance)

distance_traveled = 20
rate_per_km = 5

ola_base_price = 50
ola_fare = calculate_fare(ola_base_price, rate_per_km, distance_traveled)

uber_base_price = 35
uber_fare = calculate_fare(uber_base_price, rate_per_km, distance_traveled)

cheaper_option = "Ola" if ola_fare < uber_fare else "Uber"

print(f"Ola Fare: ${ola_fare}")
print(f"Uber Fare: ${uber_fare}")
print(f"Cheaper Option: {cheaper_option}")


```
